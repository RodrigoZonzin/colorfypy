```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

rgb_colors = [(91, 71, 83), (78, 43, 41), (76, 57, 59), (105, 33, 34)]
rgb_luminancia_ordenado = [
    (91, 71, 83),   # L ≈ 76.1
    (76, 57, 59),   # L ≈ 59.2
    (78, 43, 41),   # L ≈ 50.4
    (105, 33, 34),  # L ≈ 50.3
]
normalized_colors = [(r/255, g/255, b/255) for r, g, b in rgb_luminancia_ordenado]

cmap = LinearSegmentedColormap.from_list("custom_colormap", normalized_colors)

gradient = np.linspace(0, 1, 256).reshape(1, -1)

plt.figure(figsize=(8, 2))
plt.imshow(gradient, aspect='auto', cmap=cmap)
plt.axis('off')
plt.title("Colormap contínuo entre as cores fornecidas")
plt.show()
```
